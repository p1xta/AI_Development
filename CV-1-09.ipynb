```python
from skimage import data
import matplotlib.pyplot as plt
import cv2
```

```python
coins_image = data.coins()
```

```python
print(f"Coins image shape: {coins_image.shape}")
plt.imshow(coins_image, cmap='gray')
```

```python
image_with_edges = cv2.Canny(coins_image, threshold1=150, threshold2=200)
plt.imshow(image_with_edges, cmap='gray')
```

```python
circles = cv2.HoughCircles(image_with_edges, cv2.HOUGH_GRADIENT, dp=1, minDist=20, param1=50, param2=30, minRadius=10, maxRadius=50)
circles = circles.astype(int)
result_image = coins_image.copy()
result_image = cv2.merge([result_image, result_image, result_image])
for i in circles[0,:]:
    cv2.circle(result_image,(i[0],i[1]),i[2],(0,255,0),2)
    cv2.circle(result_image,(i[0],i[1]),2,(255,0,0),3)
print(f"{len(circles[0])} out of 24 coins found.")
plt.imshow(result_image, cmap='gray')
```
